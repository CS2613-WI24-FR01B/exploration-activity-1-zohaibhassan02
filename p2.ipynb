```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (replace 'housing.csv' with the actual filename)
housing_data = pd.read_csv('housing.csv')

# Basic data exploration
print(housing_data.head())  # Print the first few rows to understand the structure of the data
print(housing_data.describe())  # Summary statistics of numerical columns

# Data visualization using Seaborn
# Example 1: Distribution of target variable MEDV (median house value)
sns.histplot(housing_data['MEDV'], bins=20, kde=True)
plt.title('Distribution of Median House Value (MEDV)')
plt.xlabel('MEDV')
plt.ylabel('Frequency')
plt.show()

# Example 2: Scatter plot of RM (average number of rooms) vs. MEDV
sns.scatterplot(x='RM', y='MEDV', data=housing_data)
plt.title('Average Number of Rooms vs. Median House Value (MEDV)')
plt.xlabel('Average Number of Rooms (RM)')
plt.ylabel('MEDV')
plt.show()

# Example 3: Scatter plot of LSTAT (percentage of lower status population) vs. MEDV
sns.scatterplot(x='LSTAT', y='MEDV', data=housing_data)
plt.title('Percentage of Lower Status Population vs. Median House Value (MEDV)')
plt.xlabel('Percentage of Lower Status Population (LSTAT)')
plt.ylabel('MEDV')
plt.show()

# Example 4: Scatter plot of PTRATIO (pupil-teacher ratio) vs. MEDV
sns.scatterplot(x='PTRATIO', y='MEDV', data=housing_data)
plt.title('Pupil-Teacher Ratio vs. Median House Value (MEDV)')
plt.xlabel('Pupil-Teacher Ratio (PTRATIO)')
plt.ylabel('MEDV')
plt.show()

# Additional visualizations and analyses can be added as needed
```
